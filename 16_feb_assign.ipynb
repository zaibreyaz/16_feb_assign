```python
"""Q2. What is DDL? Explain why CREATE, DROP, ALTER, and TRUNCATE are used with an example.

   Ans: DDL (Data Definition Language) is a set of SQL commands used to define the structure of a database, 
        including creating, modifying, and deleting database objects such as tables, views, indexes, and constraints.

        CREATE command is used to create a new database object, such as a table or view. For example, you can create a 
        table to store customer information in a database.

        DROP command is used to delete an existing database object. For example, you can drop a table to remove it from the database.

        ALTER command is used to modify the structure of an existing database object. For example, you can alter a 
        table to add or remove columns.

        TRUNCATE command is used to remove all data from an existing table. For example, you can truncate a table 
        to remove all records from it without deleting the table itself.

        Overall, DDL commands are important for managing the structure of a database and ensuring that it meets 
        the needs of the users and applications that rely on it.
"""
```

```python
"""Q3. What is DML? Explain INSERT, UPDATE, and DELETE with an example.

    Ans: DML stands for Data Manipulation Language, which is a subset of SQL (Structured Query Language) 
         used to modify and manipulate data stored in a database.

         INSERT statement is used to add new data into a table. For example, if you have a table of employees, 
         you can use the INSERT statement to add a new employee's details like their name, email, and job title.

         UPDATE statement is used to modify existing data in a table. For instance, if an employee changes their job title, 
         you can use the UPDATE statement to modify the job title field in the table for that particular employee.

         DELETE statement is used to remove data from a table. For example, if an employee leaves the company, 
         you can use the DELETE statement to remove their data from the employee table.

         In summary, DML provides the ability to manipulate data in a database by inserting new data, updating existing data, 
         and deleting unwanted data.
"""
```

```python
"""Q4. What is DQL? Explain SELECT with an example.

    Ans: DQL stands for Data Query Language, which is a subset of SQL (Structured Query Language) used to retrieve data from a database.

         SELECT statement is used to retrieve specific data from a table in a database. For example, 
         if you have a table of employees, you can use the SELECT statement to retrieve all the 
         employees' names and job titles, or only the employees who work in a specific department.

         Here's an example of a SELECT statement:

         SELECT name, job_title
         FROM employees
         WHERE department = 'Sales'

         This statement will retrieve the names and job titles of all employees who work in the Sales department.

         In summary, DQL provides the ability to retrieve data from a database by using the SELECT statement to 
         specify the data to be retrieved and any conditions to be applied.
"""
```

```python
"""Q5. Explain Primary Key and Foreign Key.

    Ans: In a relational database, a Primary Key is a column or a combination of columns that uniquely identifies 
         each row in a table. It's like a unique identifier for each record in the table. For example, in a table of employees, 
         the primary key could be the employee ID number. Primary keys ensure that each record in the table is unique and 
         can be accessed efficiently.

         A Foreign Key is a column or a combination of columns that is used to establish a relationship between two tables. 
         It's a reference to the primary key in another table. For example, if you have a table of orders and 
         a table of customers, the foreign key in the orders table could reference the primary key of the customers table. 
         This relationship allows you to link data between the two tables and ensure that data remains consistent across them.
"""
```

```python
"""Q6. Write a python code to connect MySQL to python. Explain the cursor() and execute() method.

    Ans: To connect MySQL to Python, you need to install the MySQL Connector package using pip. 
         Here's a sample code snippet to connect to a MySQL database and execute a simple query:
"""

import mysql.connector

mydb = mysql.connector.connect(
  host="localhost",
  user="yourusername",
  password="yourpassword",
  database="yourdatabase"
)

mycursor = mydb.cursor()

mycursor.execute("SELECT * FROM yourtable")

myresult = mycursor.fetchall()

for x in myresult:
  print(x)

"""
        In this code, we first import the MySQL Connector package and then use the mysql.connector.connect() method to 
        establish a connection to the MySQL database. We pass in the host, username, password, and database name as parameters.

        Next, we create a cursor object using the mydb.cursor() method. A cursor is an object that allows you to 
        execute SQL queries on a database and fetch data from the result sets.

        We then use the mycursor.execute() method to execute a SELECT query on a table in the database. 
        The query is passed in as a string parameter.

        The mycursor.fetchall() method is used to retrieve all the rows of the result set as a list of tuples.

        Finally, we loop through the result set and print out each row using a for loop.

        In summary, the cursor() method creates a cursor object that allows us to interact with the database, 
        while the execute() method is used to execute SQL queries on the database.
"""
```
